```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("twitter_training.csv",header=None)
df.head()
```

```python
df.columns=['id','entity','sentiment','text']
df.head()
```

```python
df.columns
```

```python
df.dropna(subset=['text','sentiment'],inplace=True)
df['text']=df['text'].astype(str)
```

```python
#Sentiment Distribution Plot
sns.countplot(x='sentiment',data=df)
plt.title("Sentiment Distribution in Social Media Data")
plt.xlabel('Sentiment')
plt.ylabel("Count")
plt.show()

```

```python
#Sentiment Percentage

df['sentiment'].value_counts(normalize=True)*100
```

```python
#Entity-wise Sentiment

top_entities=df['entity'].value_counts().head(5).index
filtered_df=df[df['entity'].isin(top_entities)]

plt.figure(figsize=(8,5))
sns.countplot(x='entity',hue='sentiment',data=filtered_df)
plt.title('Sentiment by Top Entities')
plt.xticks(rotation=30)
plt.show()
```

## Insights & Observations

- Positive sentiment is dominant in the social media dataset.
- Negative sentiment highlights user dissatisfaction toward certain entities.
- Neutral sentiment represents informational or opinion-free content.
- Social media sentiment analysis helps understand public opinion effectively.
